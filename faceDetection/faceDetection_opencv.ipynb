# Face detection using openCV

### Import the required packages

```python
import cv2
import numpy as np
```

### Define the video used and the path

```python
title = "Cepages_resistants"
video_path = "\"../videos/" + title + "/" + title + ".mp4\""
```

```python
video_path = "/home/yoanwm/Documents/TER/TER-Evaluation-Transcription-automatique-d-audio/videos/Cepages_resistants/Cepages_resistants.mp4"
```

```python
cap= cv2.VideoCapture(video_path)
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
frame_width = int(cap.get(3))
frame_height = int(cap.get(4)) 
size = (frame_width, frame_height) 
result = cv2.VideoWriter('Cepages_resistants.mp4',  
                         cv2.VideoWriter_fourcc(*'MJPG'), 
                         10, size) 
while True:
    ret, frame = cap.read()
    if ret :
        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        faces = face_cascade.detectMultiScale(gray, 1.3, 5)
        for (x, y, w, h) in faces:
            cv2.rectangle(frame, (x, y), (x + w, y + h), (255, 0, 0), 5)
            roi_gray = gray[y:y+w, x:x+w]
            roi_color = frame[y:y+h, x:x+w]

        #cv2.imshow('frame', frame) #Si on veut voir le r√©sultat en direct
        result.write(frame) 
    if cv2.waitKey(1) == ord('q'):
            break

cap.release()

result.release() 
cv2.destroyAllWindows()
    
```
